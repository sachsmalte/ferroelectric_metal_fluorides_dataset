# Characterization of potential ferroelectrics via a SVM analysis
The aim of this work is to characterize polar metal fluorides into three groups:
- ferroelectric (FE)
- polar, but only pyroelectric (PE)
- false polar space group, actually centrosymmetric (WM)

The ferroelectric compounds are distinguished between already known ferroelectrics (FE_I) and possible ferroelectrics (FE_II)

We used the following parameters:
- $\Delta E$ (energy difference without lattice strain in eV per atom)
- $\Delta H$ (energy difference of two optimizations in polar and centrosymmetric structure in eV per atom)
- $d_{rms}$ (mean displacement difference from pymatgen structure similarity routine in Ang)
- $d_{max}$  (max displacement from pymatgen structure similarity routine in Ang)




## One-Class SVM

Basic idea:

The goal is to enclose the training data set in the feature space with a decision boundary so that the enclosed volume is minimal. For new data points, the distance to the decision boundary is evaluated:

- Points within the area are considered to belong the the group of the training data.

- the more central they are, the higher their agreement with the training set.

- Points outside the decision area are classified as deviating from the training set, with the degree of deviation increasing with distance.

## Assignment to Group WM: incorrect structural model

The assignment is primarily based on analysis of the DFT structural optimizations in the polar structure and centrosymmetric supergroup. Polar structures of this group converge to the centrosymmetric energy minimum during structural optimization. Due to numerical inaccuracies, we have set the following thresholds to form this category:
- $\Delta H$ < 0.1 meV **and** $d_{rms}$ < 0.01 Å **and** $d_{max}$ < 0.02 Å

21 of the 58 structure types fall into this group and are included in the current data set. Two more compounds are found to have incorrect models (Sr$_3$Fe$_2$F$_{12}$, UF$_3$). However, their DFT structural optimizations do not converge into a centrosymmetric model and are therefore excluded from the SVM-Analysis. 


## Assignment to Group PE: polar and pyroelectric

A number of polar compounds do not have a centrosymmetric supergroup. By definition, they are not ferroelectric, but potentially pyroelectric. There are 14 such compounds. They are not included in the data set, because the DFT parameters are not available for them.

## Assignment to Group FE: polar and ferroelectric

There remain 20 polar structure types for which a centrosymmetric supergroup has been found and which are potentially ferroelectric (FE). Six of these structure types  have already been characterized or discussed as ferroelectric in the literature:
- K$_2$Mn$_2$(BeF$_4$)$_3$
- BaZnF$_4$
- (NH$_4$)$_2$BeF$_4$
- Pb$_5$Al$_3$F$_19$
- LiFe$_2$F$_6$
- CsPbF$_3$

For these and known substitution variants, a total of 16 compounds, (FE_I) we calculated the above parameters and used them to train a “one-class supported vector machine” model to quantify the similarity of the remaining 14 structural types (FE_II) to these known ones.

The parameters $d_{rms}$ and $d_{max}$ are strongly correlated. Therefore, only the parameters $\Delta H$, $\Delta E$, and $d_{rms}$ are used for the SVM.

### Feature scaling
The data is scaled to a uniform value range of +- 1 using a MinMaxScaler.

```python
scale = MinMaxScaler()

scale.fit(data[["delta_h","delta_e","rms_dist"]])

data_scaled = scale.transform(data[["delta_h","delta_e","rms_dist"]])

data["delta_h_scl"] = data_scaled.T[0]
data["delta_e_scl"] = data_scaled.T[1]
data["rms_dist_scl"] = data_scaled.T[2]
```

### SVM model
Create training and test data sets, create the SVM model, and calculate the score for the test data set.

```python
train = data[data["category"] == "FE_I"]
test =  data[data["category"] == "FE_II"]

X_train = train[["delta_h_scl","delta_e_scl","rms_dist_scl"]].values
X_test = test[["delta_h_scl","delta_e_scl","rms_dist_scl"]].values
```

### Hyperparameter-Tuning of $\nu$ value:
The parameter $\nu$ of the SVM model is optimized by maximizing $\sum(s_i)/\sum(|s_i|)$, with $s_i$ beeing the score of the training data point $i$.

```python
for nu in 0.5, 0.4, 0.2, 0.1, 0.01:

    model = OneClassSVM(nu=nu,gamma="scale",kernel="rbf")
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                           
    model.fit(X_train)
    model.predict(X_train)
    scores = model.decision_function(X_train)
    print(nu, sum(scores)/sum(abs(scores)))
```

The best parameter of $\nu$ is 0.2.

```python
model = OneClassSVM(nu=0.2,gamma="scale",kernel="rbf")
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                           
model.fit(X_train)
model.predict(X_test)

scores = model.decision_function(X_test).reshape(-1,1)

test["score"] = scores
result = test[["formula","category","pointgroup","sg_number","pearson","score","delta_h","delta_e","rms_dist"]].sort_values(by="score",ascending=False)

min_val = result['score'].min()
max_val = result['score'].max()
a=0
b=1
result['score_scaled'] = round(a + (result['score'] - min_val) / (max_val - min_val) * (b-a),2)

result.sort_values(by=["score"],ascending=False)
```

According to this model, four compounds are clearly classified as potentially ferroelectric, six are close, but outside to the decision boundary and four compounds are clearly different from the training data set: Li3AlF6, BaLi2Al2F10, RbTh3F13, and KNa2ZrF7
